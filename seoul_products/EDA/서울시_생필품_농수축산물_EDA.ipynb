```python
near_markets = data[data["시장/마트 이름"].isin(["백련시장", "망원시장", "마포농수산물시장", "대림시장", "증산종합시장"])]
```

```python
# 품목별 가격 평균 계산
avg_price = near_markets.groupby(["품목 번호", "품목 이름"])["가격(원)"].mean().reset_index()
avg_price.rename(columns={"가격(원)": "평균 가격(원)"}, inplace=True)
avg_price
```

###  서울시 전체와 평균 가격 비교 (위의 시장들도 포함 - 전체 흐름 파악이 목적이니까)

```python
avg_price_tot = data.groupby(["품목 번호", "품목 이름"])["가격(원)"].mean().reset_index()
avg_price_tot.rename(columns={"가격(원)": "평균 가격(원)"}, inplace=True)
avg_price_tot
```

```python
# natural join
merged_data = pd.merge(avg_price, avg_price_tot, on=["품목 번호", "품목 이름"], suffixes=('_market', '_total'))

# 'cheaper' 컬럼 추가
merged_data['cheaper'] = (merged_data["평균 가격(원)_market"] < merged_data["평균 가격(원)_total"]).astype(int)

# '가격 차이(원)'과 '가격 차이(%)' 계산
merged_data['가격 차이(원)'] = merged_data["평균 가격(원)_total"] - merged_data["평균 가격(원)_market"]
merged_data['가격 차이(%)'] = (merged_data['가격 차이(원)'] / merged_data["평균 가격(원)_total"]) * 100

print(merged_data)
```

### 카테고리 그룹화
- 농산물: 채소, 과일, 곡물 등
- 축산물: 돼지고기, 소고기, 닭고기 등
- 생선/해산물: 조개, 새우, 오징어, 갈치 등
- 생필품: 비누, 샴푸, 칫솔, 치약 등
- 인스턴트: 라면, 즉석밥, 통조림 등
- 음료: 콜라, 사이다, 우유, 소주 등

```python
# 품목별 카테고리 지정
def assign_category(item):
    if isinstance(item, str):  # item이 문자열인지 확인
        for category, items in categories.items():
            if any(item.startswith(it) for it in items):
                return category
    return "기타"  # 지정되지 않은 항목 또는 NaN

if 'near_markets' in locals() and 'seoul_avg_data' in locals():  # 서울시 평균 데이터가 있어야 함
    # 카테고리 추가
    near_markets["품목 카테고리"] = near_markets["품목 이름"].apply(assign_category)
    seoul_avg_data["품목 카테고리"] = seoul_avg_data["품목 이름"].apply(assign_category)

    # 시장별 평균 가격 계산
    avg_price_by_category = (
        near_markets.groupby(["시장/마트 이름", "품목 카테고리"])["가격(원)"]
        .mean()
        .reset_index()
    )

    # 서울시 평균 가격 계산
    seoul_avg_by_category = (
        seoul_avg_data.groupby("품목 카테고리")["가격(원)"]
        .mean()
        .reset_index()
        .rename(columns={"가격(원)": "서울시 평균 가격(원)"})
    )

    # 시장별 데이터와 서울시 평균 데이터 병합
    comparison_data = pd.merge(
        avg_price_by_category,
        seoul_avg_by_category,
        on="품목 카테고리",
        how="left"
    )

    # 서울시 평균과의 차이 계산
    comparison_data["차이(원)"] = comparison_data["서울시 평균 가격(원)"] - comparison_data["가격(원)"]
    comparison_data["차이(%)"] = (
        comparison_data["차이(원)"] / comparison_data["서울시 평균 가격(원)"]
    ) * 100

    
```
