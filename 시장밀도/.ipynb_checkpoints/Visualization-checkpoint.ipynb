```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc

# 한글 폰트 설정
rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

# 데이터 로드
file_path = "/Users/goyeongmin/Documents/명지/3-2/BDP-gym/BDP-FINAL-24/시장밀도/MarketDensity.csv/part-00000-99b776b2-0b31-4912-a4af-e5fc930bdb54-c000.csv"
data = pd.read_csv(file_path, encoding="cp949")

# 데이터 정렬
data = data.sort_values(by="시장밀도", ascending=False)

# 상위 5개 지역 데이터 추출
top_data = data.head(5)

# 그래프 그리기
fig, axes = plt.subplots(1, 2, figsize=(14, 6))  # 1행 2열의 그래프
fig.suptitle("시장 밀도 분석", fontsize=20)

sns.barplot(x="자치구", y="시장밀도", data=data, palette="BuGn", ax=axes[0])
axes[0].set_title("전체 시장 밀도", fontsize=16)
axes[0].set_xlabel("자치구", fontsize=12)
axes[0].set_ylabel("시장 밀도", fontsize=12)
axes[0].tick_params(axis='x', rotation=45)  # x축 레이블 회전

sns.barplot(x="자치구", y="시장밀도", data=top_data, palette="BuGn", ax=axes[1])
axes[1].set_title("상위 시장 밀도", fontsize=16)
axes[1].set_xlabel("자치구", fontsize=12)
axes[1].set_ylabel("시장 밀도", fontsize=12)
axes[1].tick_params(axis='x', rotation=45)  # x축 레이블 회전

# 레이아웃 조정
plt.tight_layout(rect=[0, 0, 1, 0.95])

# 그래프를 PNG 파일로 저장
output_path = "/Users/goyeongmin/Documents/명지/3-2/BDP-gym/BDP-FINAL-24/시장밀도/Visualization.png"
plt.savefig(output_path, dpi=300, bbox_inches='tight')  # dpi=300은 고해상도 저장

plt.show()
```
