```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the uploaded CSV file
file_path = r"C:\Users\82105\Documents\consumer.csv"
data = pd.read_csv(file_path, encoding='euc-kr')

# 한글 폰트 설정 (Mac OS 또는 Windows 환경에 맞게 조정 필요)
plt.rcParams['font.family'] = 'AppleGothic' if plt.rcParams['backend'] == 'MacOSX' else 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 상위 및 하위 3개 구의 색상 설정
top_3_gu = ['강남구', '서초구', '송파구']  # 상위 3개 구
bottom_3_gu = ['도봉구', '강북구', '중랑구']  # 하위 3개 구

# 데이터 시각화
fig, ax = plt.subplots(2, 2, figsize=(15, 10))

# 행정동 코드명을 x축으로 한 각 그래프 그리기
columns = [
    ('평균_식료품_지출_비율', '평균 식료품 지출 비율'),
    ('평균_소득 금액', '평균 소득 금액'),
    ('지출_총금액', '총 지출 금액'),
    ('식료품_지출_총금액', '식료품 지출 총금액')
]

for idx, (col, title) in enumerate(columns):
    row, col_idx = divmod(idx, 2)
    colors = ['orange' if gu in top_3_gu + bottom_3_gu else 'blue' for gu in data['행정동_코드_명']]
    
    ax[row, col_idx].bar(data['행정동_코드_명'], data[col] / 1e8, color=colors)  # 단위를 100억 원으로 변경
    ax[row, col_idx].set_title(title)
    ax[row, col_idx].set_xlabel('행정동 코드명')
    ax[row, col_idx].tick_params(axis='x', rotation=90)
    
    # y축 숫자를 천억 단위로 포맷팅
    ax[row, col_idx].yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0f}억'))

# 레이아웃 정리
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the uploaded CSV file
file_path = 'consumer.csv'
data = pd.read_csv(file_path, encoding='euc-kr')

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 데이터 전처리: 컬럼명 공백 제거
data = data.rename(columns=lambda x: x.strip())

# 그래프 준비
fig, axes = plt.subplots(2, 2, figsize=(20, 15))
fig.suptitle('지역별 데이터 비교', fontsize=20)

# 평균_식료품_지출_비율
top_3_food_ratio = data.nlargest(3, '평균_식료품_지출_비율')['행정동_코드_명'].tolist()
bottom_3_food_ratio = data.nsmallest(3, '평균_식료품_지출_비율')['행정동_코드_명'].tolist()
colors_ratio = ['orange' if gu in top_3_food_ratio else 'green' if gu in bottom_3_food_ratio else 'blue' for gu in data['행정동_코드_명']]
axes[0, 0].bar(data['행정동_코드_명'], data['평균_식료품_지출_비율'], color=colors_ratio)
axes[0, 0].set_title('평균 식료품 지출 비율', fontsize=16)
axes[0, 0].set_ylabel('비율', fontsize=12)
axes[0, 0].tick_params(axis='x', rotation=90)

# 지출_총금액
top_3_expense = data.nlargest(3, '지출_총금액')['행정동_코드_명'].tolist()
bottom_3_expense = data.nsmallest(3, '지출_총금액')['행정동_코드_명'].tolist()
colors_expense = ['orange' if gu in top_3_expense else 'green' if gu in bottom_3_expense else 'blue' for gu in data['행정동_코드_명']]
axes[0, 1].bar(data['행정동_코드_명'], data['지출_총금액'] / 1e11, color=colors_expense)
axes[0, 1].set_title('지출 총금액', fontsize=16)
axes[0, 1].set_ylabel('단위: 천억 원', fontsize=12)
axes[0, 1].tick_params(axis='x', rotation=90)

# 평균_소득 금액
top_3_income = data.nlargest(3, '평균_소득 금액')['행정동_코드_명'].tolist()
bottom_3_income = data.nsmallest(3, '평균_소득 금액')['행정동_코드_명'].tolist()
colors_income = ['orange' if gu in top_3_income else 'green' if gu in bottom_3_income else 'blue' for gu in data['행정동_코드_명']]
axes[1, 0].bar(data['행정동_코드_명'], data['평균_소득 금액'] / 1e4, color=colors_income)
axes[1, 0].set_title('평균 소득 금액', fontsize=16)
axes[1, 0].set_ylabel('단위: 백만 원', fontsize=12)
axes[1, 0].tick_params(axis='x', rotation=90)

# 식료품_지출_총금액
top_3_food_expense = data.nlargest(3, '식료품_지출_총금액')['행정동_코드_명'].tolist()
bottom_3_food_expense = data.nsmallest(3, '식료품_지출_총금액')['행정동_코드_명'].tolist()
colors_food_expense = ['orange' if gu in top_3_food_expense else 'green' if gu in bottom_3_food_expense else 'blue' for gu in data['행정동_코드_명']]
axes[1, 1].bar(data['행정동_코드_명'], data['식료품_지출_총금액'] / 1e11, color=colors_food_expense)
axes[1, 1].set_title('식료품 지출 총금액', fontsize=16)
axes[1, 1].set_ylabel('단위: 천억 원', fontsize=12)
axes[1, 1].tick_params(axis='x', rotation=90)

# 레이아웃 정리
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
